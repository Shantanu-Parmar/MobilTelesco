```python
import os
import subprocess

# Define the root directory containing the runs
root_dir = '/content/MobilTelesco/26-Nov-2024/'
dss_exe_path = '/content/DeepSkyStacker/DeepSkyStacker/DeepSkyStackerCL.exe'  # Adjust path to where DeepSkyStacker is installed

# Function to list files for each run in Dark, Light, Flat, Offset
def list_run_files(run_folder):
    run_files = {'Dark': [], 'Light': [], 'Flat': [], 'Offset': []}

    for folder in ['Dark', 'Light', 'Flat', 'Offset']:
        folder_path = os.path.join(run_folder, folder)
        if os.path.exists(folder_path):
            for file_name in os.listdir(folder_path):
                if file_name.lower().endswith(('.dng', '.jpg')):  # Assuming images are in .dng or .jpg formats
                    run_files[folder].append(os.path.join(folder_path, file_name))
    return run_files

# Function to generate DSS list for a single run, including header and absolute paths
def generate_dss_file_list(run, files):
    dss_list = ["DSS file list", "CHECKED\tTYPE\tFILE"]  # Adding header

    # Get the absolute path for each image
    for category, file_list in files.items():
        for file_path in file_list:
            file_name = os.path.basename(file_path)
            file_type = category.lower()  # Type will be dark, light, flat, or offset
            dss_list.append(f"1\t{file_type}\t{file_path}")  # Absolute path

    # Save the list to a file
    list_filename = f"{run}_list.txt"
    with open(list_filename, 'w') as f:
        for line in dss_list:
            f.write(line + "\n")

    return list_filename


# Loop through runs, generate DSS lists, and run DeepSkyStackerCLI
for run in os.listdir(root_dir):
    run_folder = os.path.join(root_dir, run)
    if os.path.isdir(run_folder):
        print(f"Processing run: {run}")
        run_files = list_run_files(run_folder)

        # Generate DSS list with absolute paths
        list_filename = generate_dss_file_list(run, run_files)
        print(f"DSS list saved for run: {run} at {list_filename}")

```
