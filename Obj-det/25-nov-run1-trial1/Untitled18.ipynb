```python
import os
import subprocess

# Set paths
dataset_path = "dataset"
yaml_path = "data.yaml"
weights_path = r"yolov5s.pt"  # or another YOLOv5 model weight

# Check if weights file exists, if not, download automatically
def download_weights():
    if not os.path.exists(weights_path):
        print(f"{weights_path} not found. Downloading...")
        # You can manually download the weights or automatically let YOLOv5 handle it
        try:
            # This will invoke YOLOv5 to handle the automatic weight download
            subprocess.run(["python", "detect.py", "--weights", "yolov5s.pt", "--img", "640", "--source", "dummy.jpg"], check=True)
            print("Weights downloaded successfully.")
        except subprocess.CalledProcessError as e:
            print(f"Error downloading weights: {e}")
    else:
        print(f"{weights_path} found.")

# Function to run a command and capture output
# Function to run a command and capture output
def run_command(command):
    try:
        print(f"Running command: {' '.join(command)}")
        result = subprocess.run(command, check=True, text=True, capture_output=True, encoding='utf-8')
        print("Command executed successfully")
        print("Standard Output:\n", result.stdout)
        return result
    except subprocess.CalledProcessError as e:
        print(f"Error occurred: {e}")
        print("Standard Output:\n", e.stdout)
        print("Standard Error:\n", e.stderr)
        return e


# Train YOLOv5 model
def train_yolo():
    print("Starting YOLOv5 training...")
    download_weights()  # Ensure weights are available
    train_command = [
        'python', 'train.py',
        '--img', '1280',
        '--batch', '4',
        '--epochs', '50',
        '--data', yaml_path,
        '--weights', weights_path
    ]
    run_command(train_command)

# Evaluate YOLOv5 model
def evaluate_yolo():
    print("Starting YOLOv5 evaluation...")
    download_weights()  # Ensure weights are available
    eval_command = [
        'python', 'val.py',
        '--data', yaml_path,
        '--weights', 'runs/train/exp/weights/best.pt',
        '--img', '640'
    ]
    run_command(eval_command)

# Run inference (testing on new images)
def run_inference():
    print("Starting inference...")
    download_weights()  # Ensure weights are available
    test_images_path = r"test"
    inference_command = [
        'python', 'detect.py',
        '--weights', 'runs/train/exp/weights/best.pt',
        '--img', '640',
        '--source', test_images_path
    ]
    run_command(inference_command)

# Execute training, evaluation, and inference
if __name__ == "__main__":
    try:
        train_yolo()
        # evaluate_yolo()
        # run_inference()
    except Exception as e:
        print(f"Error in the overall process: {e}")
```
